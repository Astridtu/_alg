1.P（多項式時間）
定義：
P 是確定性圖靈機可以在多項式時間內解決的問題（或決策問題）的集合。
多項式時間意味著解決問題所需的時間受輸入大小的多項式函數的限制。
主要特點：
P 中的問題被認為是「可有效解決的」。
範例：
將清單進行排序（例如，合併排序、快速排序）。
尋找圖中的最短路徑（例如 Dijkstra 演算法）。
檢查數字是否為質數（使用 AKS 素數測試等現代演算法）。
2. NP（非確定性多項式時間）
定義：
NP 是一組可以透過確定性圖靈機在多項式時間內驗證解決方案的問題。
或者，NP 問題可以被認為是可以透過非確定性圖靈機（可以「猜測」解決方案的理論模型）在多項式時間內解決的問題。
主要特點：
NP 並不意味著「不是多項式」。相反，它代表“非確定性多項式”。
驗證比尋找 NP 問題的解決方案更容易。
範例：
數獨：給定一個填滿的網格，驗證它是否是有效的解決方案是很快的（多項式）。
哈密​​頓路徑：給定一條路徑，檢查它是否恰好訪問每個頂點一次是很快的。

4.P=NP？
大問題：
P = NP 嗎？
如果P=NP，那麼凡是能在多項式時間內驗證解的問題也能在多項式時間內解決。
如果 P ≠ NP，則 NP 中存在問題（如 NP 完全問題），這些問題本質上更難解決而不是驗證。
影響：
如果 P = NP：
像分解大整數（用於密碼學）這樣的問題將變得容易解決。
科學和工程方面的突破可能會發生，因為許多優化和搜尋問題將變得有效解決。
如果 P ≠ NP：
它證實了有些問題本質上解決起來比驗證更難。
目前狀態：
P = NP 是否成立的問題仍懸而未決，是千禧年獎問題之一。克萊數學研究所為正確的解決方案提供 100 萬美元的獎金。
P、NP 和 NP 完全之間的關係
P ⊆ NP：
P 中的每個問題也在 NP 中，因為如果你可以在多項式時間內解決問題，你也可以在多項式時間內驗證解決方案。
NP 完全 ⊆ NP：
所有 NP 完全問題都是 NP 問題，因為它們的解決方案可以在多項式時間內驗證。
P = NP（或不是）：
如果 P = NP，則每個 NP 問題，包括 NP 完全問題，都可以在多項式時間內求解。
如果 P ≠ NP，則 NP 完全問題在多項式時間內仍然無法解決。
